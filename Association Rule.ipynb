### Association Rule Mining

```python
# Split the product
def txt_split(txt):
    return txt.split(',')
```

```python
df= df['shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'].apply(txt_split)
df
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(df)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
data.replace([True,False],[1,0], inplace=True)
```

```python
data
```

```python
data.shape
```

```python
scores = apriori(data,min_support = 0.01,use_colnames=True)
scores
```

```python
rules = association_rules(scores,metric="confidence", min_threshold=0.2)
```

```python
rules
```

```python
rules.sort_values(by = 'lift', ascending=False)
```
