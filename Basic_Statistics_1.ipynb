### Standardization of Numerical Variables:

●Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

Standardization : It also known as z-score normalization, it is a technique used to transform numerical data to have a mean of 0 and a standard deviation of 1.
This process is particularly useful in machine learning and statistical analyses when the scale of variables may vary widely.

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Before standardization 
```

```python
df
```

```python
#Standardization
sc = StandardScaler()
```

```python
df.dtypes
```

```python
df.drop(columns=['Date','Day','SKU','City','BU','Brand','Model'], inplace=True)
```

### Conversion of Categorical Data into Dummy Variables:

●Objective: To transform categorical variables into a format that can be provided to ML algorithms.

```python
df=pd.read_csv(r'C:\Users\sanket\Desktop\DS ASSIGNMENT\Basic stats - 1 (3)\Basic stats - 1\sales_data_with_discounts.csv')
df
```

```python
df.head()
```

```python
#Encoding using pandas dummy function.
```

```python
df.dtypes
```

```python
df=pd.get_dummies(data=df, columns= ['Date','Day','SKU','City','BU','Brand','Model'])
df
```

```python
df=df.astype(int)
df
```
